```python
# =================================================-
#### Slide 44: Optimal Eps determination  ####

nn_model = NearestNeighbors(n_neighbors=10)
nbrs = nn_model.fit(paysim_dbscan_scaled)
distances, indices = nbrs.kneighbors(paysim_dbscan_scaled)
distances = np.mean(distances, axis=1)
distances = np.sort(distances, axis=0)
```

```python
# =================================================-
#### Slide 47: DBSCAN: optimized model  ####

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=10)
optimized_clusters = dbscan.fit_predict(paysim_dbscan_scaled)

# Check the number of clusters
unique, counts = np.unique(optimized_clusters, return_counts=True)
print(np.asarray((unique, counts)).T)
paysim_dbscan["cluster"] = optimized_clusters

paysim_dbscan.loc[paysim_dbscan["cluster"] >= 0, "cluster"] = 0
paysim_dbscan.loc[paysim_dbscan["cluster"] == -1, "cluster"] = 1
```

```python
# =================================================-
#### Slide 29: Prepare a dataset for decision tree modeling  ####

paysim = pd.read_csv("data/paysim_transactions.csv")
```

```python
paysim["type"].unique()
```

```python
# Drop columns.
paysim = paysim.drop(["step", "nameOrig", "nameDest", "isFlaggedFraud"], axis=1)
paysim.columns
```

```python
paysim = pd.get_dummies(paysim, columns=["type"], drop_first=True)
```

```python
paysim
```

```python
# =================================================-
#### Slide 31: Decision tree classification  ####

# Select predictors and target.
y = paysim["isFraud"]
X = paysim.drop(["isFraud"], axis=1)

# Build a logistic regression model.
np.random.seed(1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
dtree = DecisionTreeClassifier(max_depth=3)
dtree.fit(X_train, y_train)
```

```python
y.value_counts(normalize=True) * 100
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8.0, 2.2]
plt.rcParams['figure.dpi'] = 300
plot_tree(
    dtree,
    feature_names=X.columns.tolist(),
    class_names=["no_fraud", "fraud"],
    filled=True,
);
```

```python
# =================================================-
#### Slide 32: Predict the target  ####

dtree_y_train_pred = dtree.predict(X_train)
dtree_y_test_pred = dtree.predict(X_test)
dtree_accuracy = metrics.accuracy_score(y_test, dtree_y_test_pred)
print("Accuracy of test data:\t", dtree_accuracy)
# ROC AUC value.
roc_auc_score(y_test, dtree_y_test_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

cmd = ConfusionMatrixDisplay.from_estimator(dtree, X_test, y_test, colorbar=False)
(tn, fp), (fn, tp) = cmd.confusion_matrix
```

```python
from sklearn.metrics import RocCurveDisplay, recall_score, precision_score
plt.rcParams['figure.figsize'] = [4, 4]
RocCurveDisplay.from_estimator(dtree, X_test, y_test, plot_chance_level=True);
```

```python
from sklearn.metrics import PrecisionRecallDisplay
plt.rcParams['figure.figsize'] = [4, 4]
PrecisionRecallDisplay.from_estimator(dtree, X_test, y_test, plot_chance_level=True);
```

```python
from sklearn.calibration import CalibrationDisplay
plt.rcParams['figure.figsize'] = [4, 4]
CalibrationDisplay.from_estimator(dtree, X_test, y_test);
```

```python
### from sklearn.metrics import PrecisionRecallDisplay
plt.rcParams['figure.figsize'] = [4, 4]
prc = PrecisionRecallDisplay.from_estimator(dtree, X_test, y_test, plot_chance_level=True);
```

```python
# =================================================-
#### Slide 33: Confusion matrix of training data  ####

print("Confusion Matrix - Training Dataset")
print(
    pd.crosstab(
        y_train,
        dtree_y_train_pred,
        rownames=["True"],
        colnames=["Predicted"],
        margins=True,
    )
)
print("Percentage of accurate fraud cases is ", 133 / 153)
```

```python
# =================================================-
#### Slide 34: Confusion matrix of test data  ####

print("Confusion Matrix - Testing Dataset")
print(
    pd.crosstab(
        y_test,
        dtree_y_test_pred,
        rownames=["True"],
        colnames=["Predicted"],
        margins=True,
    )
)
print("Percentage of accurate fraud cases is", 45 / 67)
```

```python
print(metrics.classification_report(y_test, dtree_y_test_pred))
```

```python
# =================================================-
#### Slide 37: Find TPR and TNR  ####

tn, fp, fn, tp = confusion_matrix(y_test, dtree_y_test_pred).ravel()

# Find the TNR.
non_fraud_eval = tn / (tn + fp)
print(non_fraud_eval)
# Find the TPR.
fraud_eval = tp / (tp + fn)
print(fraud_eval)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(dtree_y_test_pred, y_test))
```

```python
# =================================================-
#### Slide 38: Save the metric  ####

perf_df = pd.DataFrame(
    classification_report(dtree_y_test_pred, y_test, output_dict=True)
).T.head(2).assign(
    model="dtree_no_smote",
    accuracy=accuracy_score(y_test, dtree_y_test_pred)
)
```
