```python
# =================================================-
#### Slide 26: Loading packages  ####

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_auc_score
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.linear_model import SGDOneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
from sklearn.covariance import EllipticEnvelope
```

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
# =================================================-
#### Slide 27: Directory settings  ####

# Set 'main_dir' to location of the project folder
from pathlib import Path

data_dir = Path.cwd() / "data"
```

```python
# https://github.com/ageron/handson-ml3
# https://github.com/ageron/handson-ml3/blob/main/09_unsupervised_learning.ipynb
X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
X2 = X2 + [6, -8]
X = np.r_[X1, X2]
y = np.r_[y1, y2]

gm = GaussianMixture(n_components=3, n_init=10, random_state=42)
gm.fit(X)
```

```python
dbscan = DBSCAN()
dbscan.fit(X)
```

```python
lof = LocalOutlierFactor()
lof_pred = lof.fit_predict(X)
lof_mask = lof_pred == -1
lof_anomalies = X[lof_mask]
```

```python
isof = IsolationForest(contamination=.01)
isof_pred = isof.fit_predict(X)
isof_mask = isof_pred == -1
isof_anomalies = X[isof_mask]
```

```python
np.unique(dbscan.labels_)
```

```python
anomalies_mask = dbscan.labels_ == -1
dbscan_anomalies = X[anomalies_mask]
```

```python
densities = gm.score_samples(X)
density_threshold = np.percentile(densities, 2)
anomalies = X[densities < density_threshold]
gm_anomalies = X[densities < density_threshold]
```

```python
gm_mask = densities < density_threshold
```

```python
both_mask = anomalies_mask & gm_mask
```

```python
both_anomalies = X[both_mask]
```

```python
svm = OneClassSVM(nu=.01, kernel="sigmoid", coef0=.9)
svm_pred = svm.fit_predict(X)
svm_mask = svm_pred == -1
svm_anomalies = X[svm_mask]
svm_anomalies.shape
```

```python
svm_sgd = SGDOneClassSVM(nu=.08, random_state=9)
svm_sgd_pred = svm_sgd.fit_predict(X)
svm_sgd_mask = svm_sgd_pred == -1
svm_sgd_anomalies = X[svm_sgd_mask]
svm_sgd_anomalies.shape
```

```python
elli = EllipticEnvelope(contamination=.01, random_state=9)
elli_pred = elli.fit_predict(X)
elli_mask = elli_pred == -1
elli_anomalies = X[elli_mask]
elli_anomalies.shape
```

```python
# default dbscan clusters do not look good
plt.scatter(
    X[:, 0],
    X[:, 1],
    edgecolors="k",
    c=dbscan.labels_,
);
```

```python
gm_labels = gm.predict(X)
```

```python
plt.scatter(
    X[:, 0],
    X[:, 1],
    edgecolors="k",
    c=gm_labels,
)
plt.scatter(
    isof_anomalies[:, 0],
    isof_anomalies[:, 1],
    color='yellow',
    marker='p',
    s=219,
    edgecolors="k",
    zorder=3,
    linewidths=.4,
    label="isof"
)
plt.scatter(
    lof_anomalies[:, 0],
    lof_anomalies[:, 1],
    color='darkviolet',
    marker='x',
    s=299,
    edgecolors="k",
    zorder=1,
    label="lof"
)
plt.scatter(
    gm_anomalies[:, 0],
    gm_anomalies[:, 1],
    color='r',
    marker='*',
    s=150,
    edgecolors="k",
    linewidths=.4,
    zorder=4,
    label="gm"
)
plt.scatter(
    dbscan_anomalies[:, 0],
    dbscan_anomalies[:, 1],
    color='darkorange',
    marker='o',
    s=299,
    edgecolors="k",
    linewidths=.4,
    zorder=2,
    label="dbscan"
)
plt.scatter(
    svm_anomalies[:, 0],
    svm_anomalies[:, 1],
    color='cyan',
    marker="$\u266B$",
    s=199,
    edgecolors="k",
    linewidths=.4,
    zorder=6,
    label="svm"
)
plt.scatter(
    svm_sgd_anomalies[:, 0],
    svm_sgd_anomalies[:, 1],
    color='lime',
    marker=".",
    s=119,
    edgecolors="k",
    linewidths=.4,
    zorder=7,
    label="svm_sgd"
)
plt.scatter(
    elli_anomalies[:, 0],
    elli_anomalies[:, 1],
    color='magenta',
    marker=r"$\mathcal{S}$",
    s=199,
    edgecolors="k",
    linewidths=.4,
    zorder=5,
    label="svm"
)
plt.legend(loc = "upper left");
```

```python
# =================================================-
#### Slide 28: Load the dataset  ####

from pathlib import Path

data_dir = Path.cwd() / "data"
paysim = pd.read_csv(data_dir / "paysim_transactions.csv")
paysim.head()
```

```python
# =================================================-
#### Slide 29: Prepare the dataset for modeling  ####

# Drop columns.
paysim = paysim.drop(["step", "type", "nameOrig", "nameDest", "isFlaggedFraud"], axis=1)
paysim.columns
```

```python
paysim["isFraud"].value_counts(normalize=True)
```

```python
# ================================================= -
#### Slide 30: LOF on fraud dataset  ####

train, test = train_test_split(paysim, test_size=0.30, random_state=42)
```

```python
# Split fraud vs non fraud.
non_fraud = train[train["isFraud"] == 0]
```

```python
fraud = train[train["isFraud"] == 1]
```

```python
non_fraud = non_fraud.drop("isFraud", axis=1)
```

```python
test = pd.concat([test, fraud])
```

```python
actual_test = test["isFraud"]
```

```python
# =================================================-
#### Slide 31: Create and fit LOF model  ####

lof = LocalOutlierFactor(
    n_neighbors=5, metric="manhattan", contamination=0.01088, novelty=True
)

# model fitting
lof.fit(non_fraud)
```

```python
test
```

```python
# =================================================-
#### Slide 32: Test predictions  ####

fraud_pred = lof.predict(test.iloc[:, :-1])
```

```python
np.unique(fraud_pred)
```

```python
fraud_pred = pd.Series(fraud_pred).map({1: 0, -1: 1})
fraud_pred
```

```python
cmd = metrics.ConfusionMatrixDisplay.from_predictions(actual_test, fraud_pred)
(tn, fp), (fn, tp) = cmd.confusion_matrix
```

```python
# ================================================= -
#### Slide 33: Find TPR and TNR  ####

non_fraud_eval = tn / (tn + fp)
print("tnr:", non_fraud_eval)
fraud_eval = tp / (tp + fn)
print("tpr:", fraud_eval)
```

```python
# =================================================-
#### Slide 2: Loading packages  ####

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_auc_score
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
```

```python
# =================================================-
#### Slide 5: Data: load energy consumption  ####

pjm_energy = pd.read_csv(
    data_dir / "PJME_hourly.csv", parse_dates=["Datetime"], index_col="Datetime"
)
pjm_energy.head()
```

```python
# =================================================-
#### Slide 6: Data: preprocessing  ####

pjm_energy.info()
pjm_energy = pjm_energy["2018":]
pjm_energy.shape
```

```python
# =================================================-
#### Slide 9: Create and fit LOF model: energy consumption  ####

lof_energy_model = LocalOutlierFactor(
    n_neighbors=50, metric="manhattan", contamination=0.01, novelty=False
)

pjm_energy["anomaly"] = lof_energy_model.fit_predict(
    pd.DataFrame(pjm_energy["PJME_MW"])
)
```

```python
pjm_energy["PJME_MW"].plot.box()
```

```python
pjm_energy = pjm_energy.reset_index()
```

```python
# =================================================-
#### Slide 11: LOF - visualize anomalies  ####

lower_threshold = pjm_energy["PJME_MW"].quantile(0.25)
upper_threshold = pjm_energy["PJME_MW"].quantile(0.75)
lof_anomalies = pjm_energy[pjm_energy["anomaly"] == -1]
# Upper range of anomalies
lof_upper_anomalies = lof_anomalies[lof_anomalies["PJME_MW"] > upper_threshold][
    "PJME_MW"
]

# Lower range of anomalies
lof_lower_anomalies = lof_anomalies[lof_anomalies["PJME_MW"] < lower_threshold][
    "PJME_MW"
]
```

```python
pjm_energy = pjm_energy.sort_values("Datetime")
```

```python
fig, ax = plt.subplots(figsize=(17, 6))
ax.plot(
    pjm_energy["Datetime"],
    pjm_energy["PJME_MW"],
    color="blue",
    label="Normal",
    zorder=1,
)
ax.scatter(
    lof_anomalies["Datetime"],
    lof_anomalies["PJME_MW"],
    color="red",
    edgecolor="k",
    linewidths=0.4,
    label="Anomaly",
    zorder=9,
)
plt.legend();
```

```python
plt.rcParams.update({"font.size": 20})
plt.hist(lof_lower_anomalies)
plt.xlabel("Anomaly points")
plt.ylabel("Frequency")
plt.title("Lower range of anomalies");
```

```python
plt.hist(lof_upper_anomalies)
plt.xlabel("Anomaly points")
plt.ylabel("Frequency")
plt.title("Upper range of anomalies");
```

```python
# =================================================-
#### Slide 34: Isolation forest - visualize anomalies  ####

# visualization
lower_threshold = pjm_energy["PJME_MW"].quantile(0.25)
upper_threshold = pjm_energy["PJME_MW"].quantile(0.75)
if_anomalies = pjm_energy[pjm_energy["anomaly"] == -1]

if_upper_anomalies = if_anomalies[if_anomalies["PJME_MW"] > upper_threshold]["PJME_MW"]
if_lower_anomalies = if_anomalies[if_anomalies["PJME_MW"] < lower_threshold]["PJME_MW"]
```
